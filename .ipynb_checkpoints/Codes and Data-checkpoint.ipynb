```python
sum_column = df_file["Co ppm"] + df_file["Zn ppm"]
df_file["Co ppm + Zn ppm"] = sum_column
print(df_file)
df_file.head()
```

```python
sns.boxplot(y='Co ppm + Zn ppm', x='Cumulus assemblage',
             data=df_file,
             width=0.5,
             palette="colorblind")
plt.title("Identification of Outliers")
plt.suptitle("")
plt.xlabel("Geological Setting")
plt.ylabel("Co ppm + Zn ppm")
plt.show()
```

```python
cols = ['Co ppm + Zn ppm', 'Ga ppm', 'V ppm']

for col in cols:
    df_file[col[0]] = df_file[col] * 100/ df_file[cols].sum(axis=1)
```

```python
import ternary
```

```python
#Set up the figure.
fig, tax = ternary.figure(scale=100)
fig.set_size_inches(10, 9)

#Plot points.
tax.scatter(df_file[['F', 'M', 'N']].values)

#Plot the points in groups.
for name, group in df_file.groupby('Cumulus assemblage'):

    #Note that we have to shuffle the order.
    #This will place F at the top, N on the left.
    #So the order is: right, top, left.
    points = group[['F', 'M', 'N']].values
    tax.scatter(points, marker='o', label=name)
    
plt.legend(bbox_to_anchor=(1.02, 1), loc="upper left", borderaxespad=0, title='Cumulus assemblage')
sns.set(rc = {'figure.figsize':(14,8)})

#Corner labels.
fontsize = 15
offset = 0.1
tax.top_corner_label("Cu ppm", fontsize=fontsize, offset=0.2)
tax.left_corner_label("Ga ppm", fontsize=fontsize, offset=offset)
tax.right_corner_label("Zn ppm", fontsize=fontsize, offset=offset)

#Decoration.
tax.boundary(linewidth=1)
tax.gridlines(multiple=10, color="gray")
tax.tricks(axis='lbr', linewidth=1, multiple=20)
tax.get_axes().axis('off')

tax.show()
```
